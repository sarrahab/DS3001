### Problem #1

Below I used a dataset on the quality of different types of red wine on multiple characteristics such as sugar levels, density, and pH.

Link: https://data.world/uci/wine-quality

#### Question: Is there a difference in the pH averages for poor, average, and good quality wines?

#### Pseudocode:
  - Group by poor, average, and good quality wines
  - Calculate pH Average
  - Show the values for the differences by using the poor wine's average pH as a baseline for comparison. Subtract average from poor, and good from poor

```python
# Loading and preparing the data
filename= r"C:\Users\Student\Downloads\DS-3001\data\winequality-red-ddl.csv"
data_1= pd.read_csv(filename)
```

```python
# Show the first five rows of the data
data_1.head()
```

#### Line-by-Line Solution:

```python
# Make dataframes for each of the three types of wine quality
poor_quality_df = data_1[data_1['text_rank'] == 'poor']
average_quality_df = data_1[data_1['text_rank'] == 'ave']
good_quality_df = data_1[data_1['text_rank'] == 'good']
```

```python
# Find the means for each of the three types of wine quality, and print them out
poor_quality_mean_pH = poor_quality_df['pH'].mean()
average_quality_mean_pH = average_quality_df['pH'].mean()
good_quality_mean_pH = good_quality_df['pH'].mean()

print("Average pH for Poor Quality Wines:", poor_quality_mean_pH)
print("Average pH for Average Quality Wines:", average_quality_mean_pH)
print("Average pH for Good Quality Wines:", good_quality_mean_pH)
```

```python
# Make a dataframe from scratch
data = {
    'Quality': ['Poor', 'Average', 'Good'], 'Average pH': [poor_quality_mean_pH, average_quality_mean_pH, good_quality_mean_pH], 
    'Difference': [0, average_quality_mean_pH-poor_quality_mean_pH,good_quality_mean_pH-poor_quality_mean_pH]
}

# Create a new DataFrame from the dictionary
wine_df = pd.DataFrame(data)

# Display the DataFrame
wine_df
```

#### Function:

```python
def average_pH_by_quality(wine_df):
    # Create separate DataFrames for each quality group.
    poor_quality_df = data_1[data_1['text_rank'] == 'poor']
    average_quality_df = data_1[data_1['text_rank'] == 'ave']
    good_quality_df = data_1[data_1['text_rank'] == 'good']

    # Calculate the average pH for each quality group.
    poor_quality_mean_pH = poor_quality_df['pH'].mean()
    average_quality_mean_pH = average_quality_df['pH'].mean()
    good_quality_mean_pH = good_quality_df['pH'].mean()

    # Create a new DataFrame to store the results.
    result_df = pd.DataFrame({
        'Quality': ['Poor', 'Average', 'Good'],
        'Average pH': [poor_quality_mean_pH, average_quality_mean_pH, good_quality_mean_pH],
        'Difference': [0, average_quality_mean_pH-poor_quality_mean_pH,good_quality_mean_pH-poor_quality_mean_pH]
    })

    return result_df

average_pH_by_quality(wine_df)
```

### Problem #2

Below I used a dataset on the many demographic qualities of regions/areas/countries in the world for the years 1950-2021.

#### Question: Is there a difference in the average male and female mortality between regions/areas of the world for the years 1950-2021?

#### Pseudocode (Once I create a new dataframe to only have the columns of interest, because this data is messy):
  - Group the data by area/region
  - Calculate average male and female mortality for each area
  - Create a new column "Mortality Difference" with the difference between male and female mortality rates
  - Arrange in descending order by "Mortality Difference"

```python
# Loading and preparing the data
with open(r"C:\Users\Student\Downloads\DS-3001\data\DEMOGRAPHIC_INDICATORS_COMPACT_REV1.csv", 'r', errors='replace') as file:
    data = file.read()
```

```python
# Show the first 10000 characters of the data
print(data[:10000])
```

```python
# Assuming 'data' contains CSV-like data with rows and columns
# I can specify a delimiter if it's not a comma-separated file (e.g., '\t' for tab-separated)
# Use StringIO to treat the 'data' string as a file-like object
df = pd.read_csv(StringIO(data), delimiter=',') 

# Now, 'df' is a DataFrame, and I can display it as a table
df.head(20)
```

```python
# Trying to see all the columns to see which of interest, are where
row_index = 15  # Replace with the index of the row I want to inspect

# Use .iloc[] to select the specific row and retrieve the values
row_values = df.iloc[row_index].tolist()

for i, value in enumerate(row_values, start=1):
    print(f"{i}. {value}")
```

```python
# Define the column renaming mapping so I can rename only the columns I care about
column_mapping = {
    'Unnamed: 10': 'Year',
    'Unnamed: 2': 'Area',
    'Unnamed: 61': 'Male Mortality',
    'Unnamed: 62': 'Female Mortality'
}

# Rename the columns
df.rename(columns=column_mapping, inplace=True)

# Select only the columns of interest
columns_of_interest = ['Year', 'Area', 'Male Mortality', 'Female Mortality']
df = df[columns_of_interest]
df.tail(10)
```

```python
# Check if 'Male Mortality' column contains numeric values
is_numeric_male_mortality = pd.to_numeric(df['Male Mortality'], errors='coerce').notna()

# Check if 'Female Mortality' column contains numeric values
is_numeric_female_mortality = pd.to_numeric(df['Female Mortality'], errors='coerce').notna()

# Print the results
print(f"'Male Mortality' contains numeric values: {is_numeric_male_mortality.all()}")
print(f"'Female Mortality' contains numeric values: {is_numeric_female_mortality.all()}")
```

```python
# Convert 'Male Mortality' column to numbers, replacing non-numeric values with NaN
df['Male Mortality'] = pd.to_numeric(df['Male Mortality'], errors='coerce')

# Convert 'Female Mortality' column to numbers, replacing non-numeric values with NaN
df['Female Mortality'] = pd.to_numeric(df['Female Mortality'], errors='coerce')
```

#### Line-by-Line Solution:

```python
# Group the data by 'Area'
grouped = df.groupby('Area')
```

```python
# Calculate average male and female mortality for each group, create a new dataframe called average_mortality_by_area
average_mortality_by_area = grouped[['Male Mortality', 'Female Mortality']].mean()
```

```python
# Create a new column 'Mortality Difference' with the difference between male and female mortality rates
average_mortality_by_area['Mortality Difference'] = average_mortality_by_area['Male Mortality'] - average_mortality_by_area['Female Mortality']
```

```python
# Reset the index to keep 'Area' as a regular column
average_mortality_by_area.reset_index(inplace=True)
```

```python
# Display the resulting DataFrame
average_mortality_by_area
```

```python
# Drop all NaN values from the dataframe
average_mortality_by_area.dropna(subset=['Male Mortality', 'Female Mortality'], inplace=True)
average_mortality_by_area
```

```python
# Round all numeric column values to two decimal places
average_mortality_by_area = average_mortality_by_area.round(2)
average_mortality_by_area
```

```python
# Arrange in descending order by "Mortality Difference"
average_mortality_by_area = average_mortality_by_area.sort_values(by="Mortality Difference", ascending=False)
average_mortality_by_area
```

#### Function:

```python
def process_average_mortality(df):
    # Group the data by 'Area'
    grouped = df.groupby('Area')

    # Calculate average male and female mortality for each group
    average_mortality = grouped[['Male Mortality', 'Female Mortality']].mean()

    # Create a new column 'Mortality Difference' with the difference between male and female mortality rates
    average_mortality['Mortality Difference'] = average_mortality['Male Mortality'] - average_mortality['Female Mortality']

    # Reset the index to keep 'Area' as a regular column
    average_mortality.reset_index(inplace=True)

    # Drop rows with NaN values
    average_mortality.dropna(inplace=True)

    # Round all numeric values to 2 decimal places
    average_mortality = average_mortality.round(2)

    # Arrange in descending order by "Mortality Difference"
    average_mortality = average_mortality.sort_values(by="Mortality Difference", ascending=False)

    # Return processed DataFrame
    return average_mortality

process_average_mortality(df)
```
